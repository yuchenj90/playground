```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y = [-1, 1]
y_pred = np.linspace(-2,2,100)
```

```python
x_plot = y_pred
```

```python
mis_class_loss = np.where(x_plot<0, 1, 0)
exp_loss = np.exp(-x_plot)
dev_loss = np.log(1+np.exp(-2*x_plot))
hinge_loss = np.where(1-x_plot>0, 1-x_plot, 0)
sq_error = (x_plot-1)**2
```

```python
plt.plot(x_plot, mis_class_loss, label=r'Misclassification loss: $\mathbb{1}(y*\hat{y}<0)$')
plt.plot(x_plot, exp_loss, label=r'Exponential loss: $\exp(-y\hat{y})$')
plt.plot(x_plot, dev_loss, label=r'Cross entropy: $\log(1+\exp(-2y\hat{y}))$')
plt.plot(x_plot, hinge_loss, label=r'Hinge loss: $(1-y\hat{y})^+$')
plt.plot(x_plot, sq_error, label=r'Squared error loss: ($y\hat{y}-1)^2$')
plt.legend()
#plt.legend(['Misclassification loss', 'exponential loss', 'binomial deviance', 'hinge loss', 'squared error loss'])
plt.title('Loss functions in classification, $y\in\{-1,1\}$')
plt.xlabel('y*y_pred')
plt.ylabel('loss')
```
