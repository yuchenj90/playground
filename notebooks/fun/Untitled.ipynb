```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alpha = np.array([1.0/100*i for i in range(101)])
```

```python
pg = (380-260*alpha)/(17-11*alpha)
ps = (176-80*alpha)/(17-11*alpha)
```

```python
plt.plot(alpha, pg)
plt.plot(alpha, ps)
```

```python
dg = 120000-3000*pg
ds = 20000-1250*ps
```

```python
revenue = dg*pg+ds*ps
```

```python
plt.plot(alpha, revenue)
```
